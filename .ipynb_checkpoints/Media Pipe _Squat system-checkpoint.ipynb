```python
import cv2
import mediapipe as mp
import numpy as np

mp_drawing = mp.solutions.drawing_utils
mp_pose = mp.solutions.pose
```

```python
def calculate_angle(a,b,c):
    a = np.array(a) 
    b = np.array(b) 
    c = np.array(c) 
    
    Radian = np.arctan2(b[1]-c[1], b[0]-c[0]) - np.arctan2(b[1]-a[1], b[0]-a[0])
    angle = np.abs(Radian*180.0/np.pi)
    
    if angle >180.0:
        angle = 360-angle
        
    return int(angle) 
```

# Voice Feedback System

```python
import pyttsx3
import threading
import queue

# To play audio during detection
class AudioThread(threading.Thread):
    def __init__(self, queue):
        threading.Thread.__init__(self)
        self.queue = queue
        self.start()

    def run(self):
        Audio_engine = pyttsx3.init()
        Audio_engine.startLoop(False)
        Bool_running = True
        while Bool_running:
            if self.queue.empty():
                Audio_engine.iterate()
            else:
                message = self.queue.get()
                if message == "exit":
                    Bool_running = False
                else:
                    Audio_engine.say(message)
                    
        Audio_engine.endLoop()
    
    def empty(self):
        self.queue = []
        self.end = -1

Voice_message = queue.Queue()
New_thread = AudioThread(Voice_message)
```

## Squat Detection System

```python
cap = cv2.VideoCapture(0)
# Curl counter variables
counter = 0 
stage = None
Curentangle=0
judge = True
goodsquat =0
badsqaut =0
kneeproblem = False
Hipproblem = False
Initialize = True
KneeAnglelist = []
HipAnglelist = []
Hipylist = []
Kneeylist = []
Kneexlist = []
Footxlist = []

HipDistance = 0
KneeDistance = 0
## Setup mediapipe instance
with mp_pose.Pose(min_detection_confidence=0.5, min_tracking_confidence=0.5) as pose:
    while cap.isOpened():
        if Initialize == True:
            Voice_message.put("Stand still")
            Voice_message.put("Let your shoulder keep flat")
            Voice_message.put("As you squat, keep you knee in line with your toes")
            Voice_message.put("Your hip should be lower than your knee")
            Voice_message.put("Your knee should not be beyond your toes too much")
            Voice_message.put("Now let's begin squat")
            Initialize = False
            
        ret, frame = cap.read()
        # Recolor image to RGB
        image = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
        image.flags.writeable = False
        # Make detection
        results = pose.process(image)
        # Recolor back to BGR
        image.flags.writeable = True
        image = cv2.cvtColor(image, cv2.COLOR_RGB2BGR)
        
        # Extract landmarks
        try:
            landmarks = results.pose_landmarks.landmark
            # Get coordinates
            shoulder = [landmarks[mp_pose.PoseLandmark.LEFT_SHOULDER.value].x,landmarks[mp_pose.PoseLandmark.LEFT_SHOULDER.value].y]
            hip = [landmarks[mp_pose.PoseLandmark.LEFT_HIP.value].x,landmarks[mp_pose.PoseLandmark.LEFT_HIP.value].y]
            knee = [landmarks[mp_pose.PoseLandmark.LEFT_KNEE.value].x,landmarks[mp_pose.PoseLandmark.LEFT_KNEE.value].y]
            ankle = [landmarks[mp_pose.PoseLandmark.LEFT_ANKLE.value].x,landmarks[mp_pose.PoseLandmark.LEFT_ANKLE.value].y]
            
            MidKnee = [landmarks[mp_pose.PoseLandmark.LEFT_HIP.value].x,landmarks[mp_pose.PoseLandmark.LEFT_KNEE.value].y]
            # Calculate angle
            angle1 = calculate_angle(shoulder, hip, knee) #hip angle
            angle2 = calculate_angle(hip, knee, ankle) #knee angle
            angle3 = calculate_angle(hip, knee, MidKnee)
                
            KneeAnglelist.append(angle2)
            HipAnglelist.append(angle1)
            
            hipy = landmarks[mp_pose.PoseLandmark.LEFT_HIP.value].y
            kneey = landmarks[mp_pose.PoseLandmark.LEFT_KNEE.value].y
            kneex = landmarks[mp_pose.PoseLandmark.LEFT_KNEE.value].x
            footx = landmarks[mp_pose.PoseLandmark.LEFT_FOOT_INDEX.value].x
            
            HipDistance = (kneey-hipy)
            KneeDistance = (footx-kneex)
            
            Hipylist.append(int(hipy*100))
            Kneeylist.append(int(kneey*100))    
            Kneexlist.append(int(kneex*100))
            Footxlist.append(int(footx*100))
            
            # Visualize angle
            cv2.putText(image, str(angle2), 
                           tuple(np.multiply(knee, [640, 480]).astype(int)), 
                           cv2.FONT_HERSHEY_SIMPLEX, 1.5, (255, 255, 255), 2, cv2.LINE_AA
                                )
            
            
            cv2.putText(image, str(angle1), 
                       tuple(np.multiply(hip, [640, 480]).astype(int)), 
                       cv2.FONT_HERSHEY_SIMPLEX, 1.5, (255, 255, 255), 2, cv2.LINE_AA
                            )
            
            # Curl counter logic
            if Initialize == False:
                if angle2 > 170 :
                   #HipDistance = 0
                    #KneeDistance = 0
                    if stage=='GOOD':
                        goodsquat+=1
                        print('goodsquat')
                        Voice_message.put('You finish a goodsquat,keep on')
                        kneeproblem == False
                        Hipproblem == True

                    if kneeproblem == True:
                        Voice_message.put('Your knee forward too much')
                        print('Your knee forward too much')
                        
                    if  Hipproblem == True:
                        Voice_message.put("You should lower your hip")
                        print('You should lower your hip')
                        

                    judge = True
                    stage = "UP"
                    kneeproblem = False
                    Hipproblem = False

                if angle2 <=130 and stage =='UP':
                    #stage="DOWN"
                    judge = False
                    counter +=1
                    print(counter)

                if HipDistance <= 0 and  judge == False :
                    if KneeDistance <= 0.03:
                        #print('hipy')
                        #print(hipy-kneey)

                        judge = True
                        stage ="GOOD"
                        kneeproblem = False
                        Hipproblem = False

                    else:
                        kneeproblem = True
                        stage="Your knee forward too much"

                elif HipDistance > 0 and  judge == False :
                    if KneeDistance <= 0.03:
                        Hipproblem = True
                        stage="Lower your hip"

                    else :
                        kneeproblem = True
                        stage="Knee forward too much & Lower your hip"
                     
        except:
            pass
        
        # Render curl counter
        # Setup status box
        cv2.rectangle(image, (210,0), (610,75), (255,0,0), -1)
        
        cv2.rectangle(image, (0,0), (205,40), (128,128,128), -1)
        
        cv2.rectangle(image, (0,42), (205,95), (128,128,128), -1)
        
        cv2.putText(image, 'Hip to down', (3,20), cv2.FONT_HERSHEY_SIMPLEX, 0.4, (0,0,0), 1, cv2.LINE_AA)
        
        cv2.putText(image, str(int(HipDistance*100))+' cm', (80,30), cv2.FONT_HERSHEY_SIMPLEX, 1.2, (255,255,255), 2, cv2.LINE_AA)
        
        cv2.putText(image, 'Distance beyond the toes', (3,55), cv2.FONT_HERSHEY_SIMPLEX, 0.4, (0,0,0), 1, cv2.LINE_AA)
        cv2.putText(image, str(int(KneeDistance*100))+' cm', (40,88), cv2.FONT_HERSHEY_SIMPLEX, 1.2, (255,255,255), 2, cv2.LINE_AA)
        
        
        if angle1<=180:
            cv2.ellipse(image,tuple(np.multiply(hip, [640, 480]).astype(int)),(30,30),0,180-angle3,180-angle3+angle1,(0,255,0),3)
        
        if angle2<=180:
            cv2.ellipse(image,tuple(np.multiply(knee, [640, 480]).astype(int)),(30,30),0,360-angle3,360-angle3+angle2,(0,255,0),3)

        # Rep data
        cv2.putText(image, 'NUMS', (220,12), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0,0,0), 1, cv2.LINE_AA)
        cv2.putText(image, str(counter), (220,65), cv2.FONT_HERSHEY_SIMPLEX, 1.5, (255,255,255), 2, cv2.LINE_AA)
        
        # Stage data
        cv2.putText(image, 'STATE', (310,12), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0,0,0), 1, cv2.LINE_AA)
        cv2.putText(image, stage, (300,65), cv2.FONT_HERSHEY_SIMPLEX, 1, (255,255,255), 2, cv2.LINE_AA)
        
        # Render detections
        mp_drawing.draw_landmarks(image, results.pose_landmarks, mp_pose.POSE_CONNECTIONS,
                                mp_drawing.DrawingSpec(color=(245,117,66), thickness=2, circle_radius=2), 
                                mp_drawing.DrawingSpec(color=(245,66,230), thickness=2, circle_radius=2) )               
        
        cv2.imshow('Squat Detection', image)

        if cv2.waitKey(10) & 0xFF == ord('q'):
            break

    cap.release()
    cv2.destroyAllWindows()
```
